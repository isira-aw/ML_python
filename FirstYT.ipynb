# **My first ML project**

# **load data set**

```python
import pandas as pd

# get CSV file
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')

# show data
df
```

## **Data preparation**


### **Data Separetion As X and Y**

```python
y = df['logS']

y
```

```python
x = df.drop('logS', axis=1)
# drop mean not consider that 'logs'
# axis = 1 mean drop function work whith as a columns
# when use a axis = 0 then drop function work as row
x
```

## **Data splitting**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=100)

x_train
# 80% of data to get train model
```

```python
x_test
# 20% of data to get test how model works
```

# **Model Bulding**

## **Linear Regression**

### **Train the Model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

### **Applying the Model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

### **Evaluate Model Performance**

```python
# y_train  (test training data good or bad)
```

```python
# y_lr_train_pred (test training data good or bad)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print("LR MsE (Train):",lr_train_mse)
print("LR R2 (Train):",lr_train_r2)
print("LR MsE (Test):",lr_test_mse)
print("LR R2 (Test):",lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

# **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

### Applying the model to make a prediction


```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

### **Evaluate model performace**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_lr_train_pred)
rf_train_r2 = r2_score(y_train, y_lr_train_pred)

rf_test_mse = mean_squared_error(y_test, y_lr_test_pred)
rf_test_r2 = r2_score(y_test, y_lr_test_pred)

rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

## **Model Comparison**

```python
df_models = pd.concat([lr_results, rf_results], axis=0)
```

```python
df_models.reset_index(drop=True)
```
